```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(1,10,3)
y=2*x+5
```

```python
plt.plot(x,y)
plt.show()
```

```python
x= np.arange(1,10,0.1)
y=2*x+5
plt.plot(x,y,color='r')
plt.title('Line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x= np.arange(1,10,0.1)
y1=2*x+5
y2= 3*x+10
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()
```

```python
fruit={'apple':30,'mango':45,'banana':10}
names=list(fruit.keys())
quantity=list(fruit.values())
```

```python
names,quantity
```

```python
plt.bar(names,quantity)
plt.show()
```

```python
plt.bar(names,quantity)
plt.title("Distrubution of fruits")
plt.xlabel('Fruits')
plt.ylabel('quantity')
plt.show()
```

```python
plt.bar(names,quantity,color='pink')
plt.title('Distribution of fruits')
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.show()
```
